```python
from plotkit.palettes import qualitative_palette
from plotkit.themes import set_theme
from plotkit.preview import show_palette
from plotkit.utils import save_palette

palette = qualitative_palette(6)

show_palette(palette)
save_palette(palette, "my_default")

set_theme("light")

# apply in matplotlib
import matplotlib.pyplot as plt

plt.bar(range(6), range(6), color=palette)
plt.show()
plt.close()
```

```python
from plotkit.auto import apply_auto_theme
import matplotlib.pyplot as plt
import numpy as np

# Automatically configure everything
palette = apply_auto_theme("scatter", n_colors=5)
print(palette)

x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y, color=palette[0])
plt.title("Auto-themed Scatter Plot")
plt.savefig("test_scatter.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set the generic font family
plt.rcParams['font.family'] = 'serif'

# Add your specific font to the top of the serif font list for priority
# Replace 'Times New Roman' with the exact name of your installed font
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

# Now create your plot
plt.figure()
plt.plot([1, 2, 3])
plt.title("Plot with Custom Font")
plt.show()
```
